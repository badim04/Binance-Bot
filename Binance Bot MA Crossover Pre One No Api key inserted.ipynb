```python
#Creating a function for the dataframe n min data
def getminutedata(symbol, interval, lookback):
    
    df = pd.DataFrame(client.get_historical_klines(symbol, interval, lookback + ' min ago UTC'))
    #More flexible without putting in the figures then lookback is plus dash mins ago, more flexible
    
    df = df.iloc[:,:6] #Means we need just all the rows n first 6 columns only
    
    df.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume'] #Renaming the columns
    
    df = df.set_index('Time') # Set Time column as index, not a column column
    
    df.index = pd.to_datetime(df.index, unit='ms') # Sets the time to a normal readable format
    
    df = df.astype(float) #All the values in the rows are displayed as strings by default so this converts it to workable/calculable number
    
    return df
```
