```python
import websocket, json, pprint, talib, numpy
```

```python
from binance.client import Client
from binance.enums import *
```

```python
SOCKET = "wss://stream.binance.com:9443/ws/dotusdt@kline_1m"
#wss://fstream-auth.binance.com
#Kline is candlesticks
#All these streams are gotten from the python binance websocket github respository
```

```python
RSI_PERIOD = 6
RSI_OVERBOUGHT = 70
RSI_OVERSOLD = 30
TRADE_SYMBOL = 'DOTUSD'
TRADE_QUANTITY = 0.5
```

```python
closes = [] # So as to store the closed prices as a list

in_position = False #To check if there's an open order already
```

```python
API_KEY = ''
API_SECRET = ''

client = Client(API_KEY, API_SECRET)
#client = Client(config.API_KEY, config.API_SECRET, tld='us')
#This is cuz he put the api key n secret in a config.py file so he was just calling the file, so import config then call it here
#The binance client link up to your account, api key n scret is from your acc, tld is US for Binance US
```

```python
def order(side, quantity, symbol, order_type=ORDER_TYPE_MARKET):#side is to buy n sell
    try:
        print("Sending Order")
        order = client.create_order(symbol=symbol, side=side, type=order_type, quantity=quantity)
        print(order)
    except Exception as e:
        print("an exception occured - {}".format(e))
        return False

    return True
```

```python

def on_open(ws):
    print('Connection Opened')
def on_close(ws):
    print('Connection Closed')
def on_message(ws, message):
    global closes, in_position #To reference the closed list
    print('Message Recieved')
    #print(message) #This prints the actual message and not just message received
    #json_message = json.loads(message)#It would convert it to a python data structure that we can use
    #pprint.pprint(json_message)#This would print it out in a better format, like up to down
    
    candle = json_message['k']
    #use k since we just wnna reference the 'k' part n not everything in the message
    
    is_candle_closed = candle['x'] # 'x' being the true or false of it's the closing price 
    close = candle['c']# 'c' is the price for that candle
    
    if is_candle_closed:
        print("Candle Closed at {}".format(close))
        closes.append(float(close))# This adds the previous close to the next one, dont forget as a list, (above)
        #Float to convert fron just a string to a number we can numpy work on
        print("closes")
        print(closes)#This starts to accumulate over time in a list
    
        if len(closes) > RSI_PERIOD:#If the amount of rsi's are more than 14(RSI_PERIOD value)
            np_closes = numpy.array(closes)#Convert closes to a numpy array
            rsi = talib.RSI(np_closes, RSI_PERIOD)#Use talib formula to get RSI formula for the closes and PERIOD
            print('All RSIs so far')
            print(rsi)#This will print a series(list) of RSIs from the 15th upwards, 14 is the period its on now
            last_rsi = rsi[-1]#This prints the last RSI, that's (-1)
            print('The Current RSI is {}'.format(last_rsi))
            
            if last_rsi > RSI_OVERBOUGHT:#If the last rsi is more than the RSI_OVERBOUGHT(80) then print sell
                if in_position:
                    print('SELL')
                    order_succeeded = order(SIDE_SELL, TRADE_QUANTITY, TRADE_SYMBOL)
                    #order, side is side sell(sell), quantity is trade quantity, symbol, same, then order type is default
                    if order_succeeded:
                        in_position = False # So it doesnt sell more than once
                else:
                    print('OVERBOUGHT BUT NO POSTION')
                
            if last_rsi < RSI_OVERSOLD:#If the last rsi is less than the RSI_OVERSOLD(20) then print buy
                if in_position:
                    print('OVERSOLD ALREADY, OPEN POSITION')
                else:
                    print('BUY')
                    order_succeeded = order(SIDE_BUY, TRADE_QUANTITY, TRADE_SYMBOL)
                    if order_succeeded:
                        in_position = True #this is to end the trade
```

```python
ws = websocket.WebSocketApp(SOCKET, on_open=on_open, on_close=on_close, on_message=on_message)
#When connected to the ws it should runs these commands accordingly
```

```python
ws.run_forever() #This the websocket should continue running
```
